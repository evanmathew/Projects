```python
# Importing selenium and Chromedriver_binary
from selenium import webdriver
import chromedriver_binary

# Importing BeautifulSoup for parsing HTML document from webpage
from bs4 import BeautifulSoup

# Importing Data Processing and Lninear Algebra Module
import pandas as pd
import numpy as np

#Importing a module which visualize us about the progress of loop iteration
from tqdm import tqdm

#importing time Module
import time
```

```python
# Initiating the chrome driver in the system ,which will will pop up as an individual chrome testing browser as
browser = webdriver.Chrome() #(OPTIONAL) You can insert your chrome driver.exe path as a parameter

# Providing the youtube channel link (FORMAT : https://www.youtube.com/@CHANNEL_NAME/videos ) 
browser.get('https://www.youtube.com/@GeeksforGeeksVideos/videos')
time.sleep(3) # Giving some time to load whole webpage 

soup = BeautifulSoup(browser.page_source,'html.parser') #Parsing the HTML code
```

```python

data = [] # We will be storing all the data as a list 

for sp in soup.find_all('ytd-rich-item-renderer'):
    
    title       = sp.find('a',class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    video_link  = sp.find('a',class_ = 'yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    
    try:
        views       = sp.find_all('span',class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[0].text
    except:
        view = None
    
    try:
        date_time   = sp.find_all('span',class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[1].text
    except:
        date_time = None
        
    try:
        thumbnail = sp.find('img').get('src').split('?')[0]
    except:
        thumbnail = None
        
    #Appending new scrapped data into current data list with old data
    data.append([title,views,date_time,video_link,thumbnail])
    
    
    #print(title)
    #print(video_link)
    #print(dbate_time)
    #print(thumbnail)
    #print('=======================================')
    
```

```python
# Creating dataframe from the data list 
df = pd.DataFrame(data,columns=['title','views','date_time','video_link','thumbnail_link'])
df.head()
```

#### Scapping more information of every videos such as title, views, likes, date of published, and the video description.

```python

data = []

for link in df['video_link']: # loop to extract every video link from dataframe and run in chrome driver
    link = 'https://www.youtube.com' + link # Final Link
    browser.get(link) #Chrome Driver executes
    
    time.sleep(3) # Give time to load up the whole webpage before scrapping
    
    soup = BeautifulSoup(browser.page_source,'html.parser') # Parsing the HTML code
    
    
    # Try and Except method
    try:
        title     = soup.find_all('yt-formatted-string',class_ = 'style-scope ytd-watch-metadata')[0].text
    except:
        title = np.nan
    
    try:
        view      = soup.find_all('span',class_ = 'bold style-scope yt-formatted-string')[0].text
    except:
        view      = np.nan
        
    try:
         like      = soup.find_all('span',class_ = 'yt-core-attributed-string yt-core-attributed-string--white-space-no-wrap')[6].text
    except:
        like       = np.nan
    
    try:
        date_time = soup.find_all('span',class_ = 'bold style-scope yt-formatted-string')[2].text
    except:
        date_time = np.nan
    
    try:
        description = soup.find('span',class_ = 'yt-core-attributed-string yt-core-attributed-string--white-space-pre-wrap').text
    except:
        description = np.nan
    
    
    # Appending new scrapped data into current data list with old data
    data.append([title,date_time,view,like,description])
    
    # NOTE : It will take some time to scrap lot of data from these video link
    
```

```python
df = pd.DataFrame(data,columns=['Title','Date_Time','Views','Likes','Description'])
df.head()
```
