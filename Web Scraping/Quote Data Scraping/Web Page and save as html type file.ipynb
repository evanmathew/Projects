## Converting the data into csv data frame

```python
data = []
for div in soup.find_all('div',class_ = 'quote'):
    
    quote       = div.find('span',class_ = 'text').text[1:-1]
    author_name = div.find('small',class_ = 'author').text
    author_id   = div.find('a').get('href')
    
    tags=[]
    for tag in div.find_all('a',class_ = 'tag'):
        tags.append(tag.text)
    
    tags = ','.join(tags)
    quote_tags = tags
    

    author_link = 'https://quotes.toscrape.com' + author_id    
    data.append([quote,author_name,author_id,quote_tags,author_link])
    
```

```python
df = pd.DataFrame(data,columns=['Quote','Author Name','Author-ID','Tags','Author-link'])
df.head()
```

```python
df.to_csv('Quotes.csv',index=False)
```

```python
for author in df.values:
    link = author[1]
    name = 'Authors/' + author[0] + '.txt'
    res = requests.get(link)
    soup = BeautifulSoup(res.text,'html.parser')
    txt = soup.find('div',class_ = 'author-details').text[:]
    fd = open(name,'w')
    fd.write(txt)
    fd.close()
```
